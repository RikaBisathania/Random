```python
from sklearn import svm
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Membuat dataset sintetis untuk contoh
X, y = make_classification(n_samples=20000, n_features = 15)
#Tujuan dari menggunakan dataset sintetis adalah untuk melakukan eksperimen atau pengujian 
#pada algoritma atau model tanpa harus bergantung pada data nyata.

# Membagi dataset menjadi subset pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# Membuat objek model SVM
model = svm.SVC()

# Melatih model menggunakan subset pelatihan
model.fit(X_train, y_train)

# Memprediksi kelas target pada subset pengujian
y_pred = model.predict(X_test)

# Menghitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model SVM:", accuracy)
```
